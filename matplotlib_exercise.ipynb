```python
# データの読み込み
file_path = r'C:\Users\piaoh\OneDrive\デスクトップ\PG\BasicPython\data\hist_data.csv'
data = pd.read_csv(file_path)

# グラフの描画
plt.hist(data['aデータ'], bins=20, alpha=0.5, label='aデータ', color='orange')
plt.hist(data['bデータ'], bins=20, alpha=0.5, label='bデータ', color='green')
plt.xlabel('温度(°C)')
plt.ylabel('頻度')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
#データの読み込み
file_path = r'C:\Users\piaoh\OneDrive\デスクトップ\PG\BasicPython\data\scatter_data.csv'
data = pd.read_csv(file_path)

# データの確認
print(data.head())

# 散布図を作成
plt.figure(figsize=(8, 6))

# クラスごとに異なる色でプロット
for class_value, color in zip([0, 1, 2], ['red', 'blue', 'green']):
    subset = data[data['class'] == class_value]
    plt.scatter(subset['length (cm)'], subset['width (cm)'], label=f'class{class_value}', color=color)

# ラベルとタイトルを追加
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.legend()

# グリッドを表示
plt.grid(True)

# グラフを表示
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
# データの読み込み
file_path = r'C:\Users\piaoh\OneDrive\デスクトップ\PG\BasicPython\data\line_data.csv'
data = pd.read_csv(file_path)

# 列名を確認して適切な列名を使用
print(data.columns) 

# グラフの描画
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# 1つ目のグラフ: 折れ線グラフ
ax[0].plot(data['期間'], data['件数'], marker='o')  
ax[0].set_xlabel('期間')
ax[0].set_ylabel('件数')
ax[0].set_title('期間ごとの件数の推移')
ax[0].grid(True)  # グリッド線を追加
ax[0].tick_params(axis='x', rotation=45)

# 2つ目のグラフ: 棒グラフ
ax[1].bar(data['期間'], data['費用'])  
ax[1].set_xlabel('期間')
ax[1].set_ylabel('費用(円)')
ax[1].set_title('期間ごとの費用の推移')
ax[1].grid(True)  
ax[1].tick_params(axis='x', rotation=45)

# グラフの表示
plt.tight_layout()
plt.show()
```
